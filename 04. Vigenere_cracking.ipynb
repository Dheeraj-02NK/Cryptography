```python
Alpahbets = ' A!1B@2C#D3$E%4F^G5&H*6I(J7)K_8L-M9+N=0O:P;Q\'R"S?T/U>V<W.X,Y`Z~'

def vigenere_encrypt(plain_text,key):
    plain_text = plain_text.upper()
    key = key.upper()
    cipher_text = ''
    Key_Index = 0
    for char in plain_text:
        index = (Alpahbets.find(char) + Alpahbets.find(key[Key_Index])) % len(Alpahbets)
        cipher_text = cipher_text + Alpahbets[index]
        Key_Index = Key_Index + 1
        if Key_Index == len(key):
            Key_Index = 0
    return cipher_text

def vigenere_decrypt(cipher_text,key):
    cipher_text = cipher_text.upper()
    key = key.upper()
    plain_text = ''
    Key_Index = 0
    for char in cipher_text:
        index = (Alpahbets.find(char) - Alpahbets.find(key[Key_Index])) % len(Alpahbets)
        plain_text = plain_text + Alpahbets[index]
        Key_Index = Key_Index + 1
        if Key_Index == len(key):
            Key_Index = 0
    return plain_text

def kasiski(ciphertext):
    distances = {}
    for i in range(len(ciphertext)):
        for j in range(i+3, len(ciphertext)):
            if ciphertext[i:i+3] == ciphertext[j:j+3]:
                distance = j - i
                if distance not in distances:
                    distances[distance] = 1
                else:
                    distances[distance] += 1
    return distances

if __name__ == '__main__':
    text = input("Enter the message: ")
    KEY = input("Enter the key: ")
    encrypted_msg = vigenere_encrypt(text,KEY)
    print('Encrypted message is :',encrypted_msg)

    distances = kasiski(encrypted_msg)

    print("Possible key lengths:")
    for key_length in distances.keys():
        if distances[key_length] > 1:
            print(key_length)

    key_length = int(input("Enter the key length: "))
    subtexts = []
    for i in range(key_length):
        subtext = ""
        for j in range(i, len(encrypted_msg), key_length):
            subtext += encrypted_msg[j]
        subtexts.append(subtext)

    key = ""
    for subtext in subtexts:
        freqs = [0]*len(Alpahbets)
        for char in subtext:
            freqs[Alpahbets.find(char)] += 1
        max_index = freqs.index(max(freqs))
        key += Alpahbets[max_index - Alpahbets.find('E')]

    print("Key:", key)
    decrypted_msg = vigenere_decrypt(encrypted_msg,key)
    print('Decrypted message is :',decrypted_msg)
```
