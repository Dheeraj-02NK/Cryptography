# Brute Force Attack

```python
Alpahbets = ' A!1B@2C#D3$E%4F^G5&H*6I(J7)K_8L-M9+N=0O:P;Q\'R"S?T/U>V<W.X,Y`Z~'

def crack_caesar(cipher_text):
    for key in range(len(Alpahbets)):
        plain_text = ''
        
        for c in cipher_text:
            index = Alpahbets.find(c)
            index = (index - key) % len(Alpahbets)
            plain_text = plain_text + Alpahbets[index]
        print('For key %s => the result is %s\n' %(key,plain_text))
          
if __name__ == '__main__':
    encrypted_msg = input("Enter the Ecrypted message: ")
    print('\n\n')
    crack_caesar(encrypted_msg)
    
```

# Frequency Analysis

```python
import matplotlib.pylab as plt

Alpahbets = ' A!1B@2C#D3$E%4F^G5&H*6I(J7)K_8L-M9+N=0O:P;Q\'R"S?T/U>V<W.X,Y`Z~'

def frequency_analysis(cipher_text):
    cipher_text = cipher_text.upper()
    letter_frequency = {}        # we can store key value pair
    
    for letter in Alpahbets:
        letter_frequency[letter] = 0
        
    for letter in cipher_text:
        if letter in Alpahbets:
            letter_frequency[letter] += 1
    
    return letter_frequency

def plot(frequencies):
    plt.bar(frequencies.keys(),frequencies.values())
    plt.show()
    
def crack_fa(cipher_text):
    freq = frequency_analysis(cipher_text)
    freq = sorted(freq.items(),key = lambda x: x[1],reverse=True)

    Key = Alpahbets.find(freq[0][0])-Alpahbets.find(' ')
    print("The possible keys values:",Key)
    
if __name__ == '__main__':    
    text = input("Enter the text: ")
    crack_fa(text)
    
```
